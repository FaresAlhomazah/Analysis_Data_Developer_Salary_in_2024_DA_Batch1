##  Introduction

The 2024 Data Developer Salary and Job Descriptions profile provides a comprehensive look at the changing data developer work environment. This profile includes key variables such as:

- `experience_level`: Level of professional experience (e.g., junior, mid, senior).

- `employment_type`: Type of job contract (e.g., full-time, part-time, contract).

- `job_title`:The specific role or title of the employee (e.g., Data Engineer).

- `salary`: The compensation received, in the original currency.

- `salary_currency`:The currency in which the salary is paid.

- `salary_in_usd`:The salary converted into US dollars for comparison.

- `employee_residence`: The location where the employee resides.

- `remote_ratio`: Percentage of work done remotely.

- `company_location`: The geographical location of the company.

- `company_size`: The scale of the company, often based on employee count.


## More Details

### 1. `work_year`  
- **Description**: The year in which the data was recorded.  
- **Possible Values**: Contains only `2024` as the dataset pertains to this year.  
- **Meaning**: Indicates the current year during which the employees are working.  
- **Type**: Numeric (Integer).  

---

### 2. `experience_level`  
- **Description**: The professional experience level of the employee.  
- **Possible Values**:  
  - `EN`: Entry-Level.  
  - `MI`: Mid-Level.  
  - `SE`: Senior-Level.  
  - `EX`: Expert-Level.  
- **Meaning**: Represents the level of expertise of the employee in their field, which significantly impacts salary.  
- **Type**: Categorical.  

---

### 3. `employment_type`  
- **Description**: Type of job contract.  
- **Possible Values**:  
  - `FT`: Full-Time.  
  - `PT`: Part-Time.  
  - `CT`: Contract.  
  - `FL`: Freelance.  
- **Meaning**: Indicates whether the job is full-time, part-time, or another arrangement.  
- **Type**: Categorical.  

---

### 4. `job_title`  
- **Description**: The specific role or title of the employee.  
- **Possible Values**: Examples include:  
  - `AI Engineer`.  
  - `Data Engineer`.  
  - `Machine Learning Engineer`.  
  - `NLP Engineer`.  
  - `Data Analyst`.  
- **Meaning**: Refers to the employee's job title, which often correlates with their salary.  
- **Type**: Text.  

---

### 5. `salary`  
- **Description**: The compensation received in the original currency.  
- **Possible Values**: Numeric values such as `202730`, `96000`.  
- **Meaning**: Reflects the raw salary of the employee before currency conversion.  
- **Type**: Numeric (Integer).  

---

### 6. `salary_currency`  
- **Description**: The currency in which the salary is paid.  
- **Possible Values**: `USD` (United States Dollar).  
- **Meaning**: Denotes the currency used to pay the salary. In this dataset, all salaries are in USD.  
- **Type**: Text.  

---

### 7. `salary_in_usd`  
- **Description**: The salary converted into US dollars for standardized comparison.  
- **Possible Values**: Numeric values such as `202730`, `96000`.  
- **Meaning**: Enables direct comparison of salaries, unaffected by currency differences.  
- **Type**: Numeric (Integer).  

---

### 8. `employee_residence`  
- **Description**: The location where the employee resides.  
- **Possible Values**: Country codes like `US` (United States).  
- **Meaning**: Indicates the employeeâ€™s residence, which may impact the nature of contracts and salary.  
- **Type**: Text.  

---

### 9. `remote_ratio`  
- **Description**: The percentage of work completed remotely.  
- **Possible Values**:  
  - `0`: Fully onsite.  
  - `50`: Hybrid work.  
  - `100`: Fully remote.  
- **Meaning**: Specifies the extent of remote work associated with the job.  
- **Type**: Numeric (Integer).  

---

### 10. `company_location`  
- **Description**: The geographical location of the company.  
- **Possible Values**: Country codes like `US` (United States).  
- **Meaning**: Represents the company's headquarters, which can influence policies, salaries, and taxes.  
- **Type**: Text.  

---

### 11. `company_size`  
- **Description**: The scale of the company, often based on the number of employees.  
- **Possible Values**:  
  - `S`: Small.  
  - `M`: Medium.  
  - `L`: Large.  
- **Meaning**: Indicates the company size, which may impact salary and employment policies.  
- **Type**: Categorical.  

---

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("DatasetSalary2024.csv")
data.head(10)
```

```python
print("Missing values ")
print(data.isnull().sum())
```

### No Missing Values


```python
print(data['salary_in_usd'].max())
print(data['salary_in_usd'].min())
salary_min, salary_max = 15000, 500000
data = data[(data['salary_in_usd'] >= salary_min) & (data['salary_in_usd'] <= salary_max)]
```

```python
print(data.describe())
```

```python
print(data.info())
```

```python
print(data.columns)
```

```python
print(data["work_year"].unique())
print(data["experience_level"].unique())
print(data["employment_type"].unique())
uniqe_job_title=data["job_title"].unique()
print(len(uniqe_job_title))

```

### filtering data and deleting duplicated data

```python
duplicate_rows = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
# duplicate_data = data[data.duplicated()]
# print(duplicate_data)
print("Before:",len(data))
data = data.drop_duplicates()
print("After:", len(data))

```

```python
experience_salary = data.groupby('experience_level')['salary_in_usd'].mean().sort_values(ascending=False)
plt.figure(figsize=(8,6))
experience_salary.plot(kind='bar', color='skyblue')
plt.title('Average Salary by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')
plt.show()
```

### Observation:
The average salaries vary significantly by experience level. Employees at the "Expert Level"  earn the highest salaries, followed by "Senior," "Mid-level," and finally "Entry-level" with the lowest salaries.


```python
job_salary = data.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 8))
job_salary.plot(kind='barh', color='orange')
plt.title('Top 10 Jobs with Highest Average Salary')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')
plt.show()
```

### Observation:
Jobs with advanced managerial or technical titles, such as "Analytics Engineering Manager" and "Data Science Tech Lead," have the highest average salaries. 
Because these roles require extensive experience in advanced areas such as artificial intelligence, cloud computing, and data infrastructure, which are areas with high demand but a limited supply of qualified competencies. Additionally, these jobs have a significant impact on business decisions.


```python

residence_salary = data.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).head(30)
plt.figure(figsize=(10, 8))
# print(residence_salary)
residence_salary.plot(kind='barh', color='purple')
plt.title('Top 30 Employee Residences by Average Salary')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Residence')
plt.show()
```

Countries like Qatar,  and Malaysia offer the highest average salaries, while others like Sweden and the Dominican Republic have lower averages


```python

plt.figure(figsize=(8, 6))
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=data, palette='cool',hue='experience_level')
plt.title('Effect of Remote Ratio on Salary')
plt.xlabel('Remote Ratio')
plt.ylabel('Salary (USD)')
plt.show()
```


1. **Positive correlation**: Salary increases with remote work ratio.
2. **Experience impact**: Higher experience levels correspond to higher salaries.

```python
company_size_salary = data.groupby('company_size')['salary_in_usd'].mean().sort_values()
plt.figure(figsize=(8, 6))
company_size_salary.plot(kind='bar', color='green')
plt.title('Average Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Salary (USD)')
plt.show()


sns.lineplot(x='company_size', y='salary_in_usd', data=data)
plt.title('Average Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Salary (USD)')
plt.show()
```

Explains how company size affects salaries:
Medium companies pay the highest average salaries of large companies pay average salaries while small companies pay the lowest average salaries

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y='salary_in_usd', data=data, palette='Set2')
plt.title('Salary Distribution')
plt.ylabel('Salary (USD)')
plt.show()

```

Shows outliers in salaries and average salaries between 1,000,000 and 200,000

```python

plt.figure(figsize=(8, 6))
sns.kdeplot(data['salary_in_usd'], shade=True, color='red')
plt.title('Salary Density Plot')
plt.xlabel('Salary (USD)')
plt.show()
```

Shows the general distribution of salaries 

```python

most_common_jobs = data['job_title'].value_counts().head(10)
plt.figure(figsize=(10, 6))
most_common_jobs.plot(kind='bar', color='teal')
plt.title('Top 10 Most Common Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.show()
```

Displays the most popular and in-demand jobs such as Data Engineer, Data Scientist and Data Analyst

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', hue='company_size', data=data, palette='dark')
plt.title('Salary by Experience Level and Company Size')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.legend(title='Company Size')
plt.show()


plt.figure(figsize=(8, 6))
sns.scatterplot(x='experience_level', y='salary_in_usd', data=data, hue='company_size', palette='viridis', s=100)
plt.title('Relationship Between Experience Level and Salary')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.legend(title='Company Size')
plt.show()
```

We notice an average salary higher than 200,000 to 300,000 with large and medium companies and with increasing experience and the lowest average salary from 50,000 to 100,000 with small and medium companies for beginners 

```python


plt.figure(figsize=(8, 6))
sns.scatterplot(x='remote_ratio', y='salary_in_usd', size='salary_in_usd', 
                hue='employment_type', data=data, sizes=(20, 200), palette='dark')
plt.title('Effect of Remote Ratio on Salary')
plt.xlabel('Remote Ratio')
plt.ylabel('Salary (USD)')
plt.legend(title='Employment Type')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=data, hue='employment_type', palette='dark')
plt.title('Effect of Remote Ratio on Salary')
plt.xlabel('Remote Ratio')
plt.ylabel('Salary (USD)')
plt.legend(title='Employment Type')
plt.show()
```

Observation is the highest salary for those who work Fully onsite and Fully remote in full-time. 

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['employee_residence'].head(5000), y='salary_in_usd', data=data, hue='experience_level', palette='plasma', s=80)
plt.title('Relationship Between Residence and Salary')
plt.xlabel('Employee Residence')
plt.ylabel('Salary (USD)')
plt.xticks(fontsize=10,rotation=90)
plt.show()
print()


```

The analysis highlights significant variations in salaries across employee residences, with notable differences based on experience levels. Countries with higher average salaries often correlate with higher experience levels.
